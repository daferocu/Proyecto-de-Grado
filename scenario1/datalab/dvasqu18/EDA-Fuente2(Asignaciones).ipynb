##### Creamos un indice de NIT para asignar el nombre de la entidad matriz a cada registro

```python
indice_nit = pd.read_csv('index_nit.csv', encoding= 'unicode_escape')
indice_nit.head()
```

Aseguramos que la variable "NIT" conserve el mismo formato en ambos lados y modificamos el nombre de entidad por "entidad_matriz"

```python
#Renombrar variables
indice_nit  = indice_nit.rename(columns={'NIT': "nit",
                                        'entidad': "entidad_matriz"}) 
indice_nit
```

Agregamos un merge entre la matriz de entidades y el dataframe.

```python
from pandas import DataFrame, merge
```

```python
df = indice_nit.merge(df, on='nit', how='left')
df
```

```python
#Generar tabla dinamica, dejando NIT como valor del index

df_pivot = pd.pivot_table(df,index=["entidad_matriz","entidad"],values=["nit"])
print(df_pivot)
df_pivot.dtypes
```

```python
# Select all duplicate rows based on one column
duplicateRowsDF = df_pivot[df_pivot.duplicated(['nit'])]
print("Duplicate Rows based on a single column are:", duplicateRowsDF, sep='\n')
```
