# Análisis Multivariante

```python
# Tabla de contingencia modalidad / year
pd.crosstab(index=df['modalidad'],
            columns=df['year'], margins=True)
```

```python
# Gráfico de barras de modalidad segun familia
plot = pd.crosstab(index=df['nombre_familia'],
            columns=df['modalidad']).apply(lambda r: r/r.sum() *100,
                                              axis=1).plot(kind='bar')
```
